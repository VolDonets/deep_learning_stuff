```python
from keras_preprocessing.image import load_img, img_to_array, array_to_img, ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
)
```

```python
img = load_img("../src/images/test_img.png",
                     target_size=(150, 150))
x = img_to_array(img)   # convert to (150, 150, 3) tensor
x = x.reshape((1,) + x.shape)   # reshape to (1, 150, 150, 3)

i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break
plt.show()
```
